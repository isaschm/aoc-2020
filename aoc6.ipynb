```python
answers = []
with open('input.txt','r') as input_file:
    data_str = input_file.read()
    data_array = data_str.split('\n\n')
    for i, smaller_data in enumerate(data_array):
        answers.append(smaller_data)
```

```python
sum = 0
for i in range(len(answers)):
    questions = answers[i].replace("\n", "")
    yes = []
    for j in range(len(questions)):
        if questions[j] not in yes:
            yes.append(questions[j])
    sum += len(yes)
print(sum)
```

```python
total = 0
for i in range(len(answers)):
    questions = [set(item) for item in answers[i].strip().split("\n")]
    u = set.intersection(*questions)
    total += len(u)

print(total)
```
